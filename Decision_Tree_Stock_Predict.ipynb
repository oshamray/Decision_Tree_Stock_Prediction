```
#import the lib

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import yfinance as yf
import numpy as np
import pandas as pd
```

```
df = yf.download("AMD", start="2020-08-17", end="2020-09-17")
df
```

```
# Create target column
df['Price_Up'] = np.where(df['Close'].shift(-1)>df['Close'],1,0)
df
```

```
# Split the data set into a feature and target data set
X = df.iloc[:,0:df.shape[1]-1].values
Y = df.iloc[:,df.shape[1]-1].values
```

```
#Split data to 80% training and 20% testing data sets
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2)
```

```
#Create and train a module (DesicionTree)
tree = DecisionTreeClassifier().fit(X_train, Y_train)

#Show the model tested 
print(tree.score(X_test,Y_test))
```

```
#show the model predictions 
tree_predict = tree.predict(X_test)
print(tree_predict)
```
